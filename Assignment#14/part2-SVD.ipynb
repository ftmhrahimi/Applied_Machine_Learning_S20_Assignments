## Select the red, green, and blue color channels

```python
import cv2
import numpy as np
src = cv2.imread("Penguins.jpg", cv2.IMREAD_UNCHANGED)
print(src.shape)
blue_channel = src[:,:,0]
green_channel = src[:,:,1]
red_channel = src[:,:,2]
blue_img = np.zeros(src.shape)
green_img = np.zeros(src.shape)
red_img = np.zeros(src.shape)
blue_img[:,:,0] = blue_channel
green_img[:,:,1] = green_channel
red_img[:,:,2] = red_channel
cv2.imwrite('blue-channel.png',blue_img)
cv2.imwrite('green-channel.png',green_img)
cv2.imwrite('red-channel.png',red_img)
img1 = cv2.imread('blue-channel.png')
img2 = cv2.imread('green-channel.png')
img3 = cv2.imread('red-channel.png')
```

## SVD

```python
from PIL import Image
img = Image.open('Penguins.jpg')
rgb_img = np.array(img)
img = np.sum(rgb_img, axis=-1)
```

```python
u, sigma, vt = np.linalg.svd(img)
#sigma
```

```python
compressed_img = (
    sigma[0] * np.outer(u[:, 0], vt[0])
    )
```

```python
plt.imshow(compressed_img)
```
